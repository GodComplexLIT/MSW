```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("netflix.csv", encoding='latin1')
```

1. **Distribuce hodnocení**

```python
vekove_hodnoceni = df['rating'].value_counts()
plt.figure(figsize=(10,6))
plt.pie(vekove_hodnoceni.values, labels=vekove_hodnoceni.index, autopct='%1.1f%%')
plt.title('Věkové hodnocení')
plt.show()
```

2. **Žánry**

```python
zanry = df['listed_in'].value_counts().head(30)
plt.figure(figsize=(12,6))
sns.barplot(x=zanry.index, y=zanry.values)
plt.title("Žánry")
plt.ylabel("Počet")
plt.xlabel("Žánr")
plt.xticks(rotation=90)
plt.show()
```

3. **Produkce dle země**

```python
top_countries = df['country'].value_counts().head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=top_countries.index, y=top_countries.values)
plt.title('Top země produkce')
plt.ylabel("Počet")
plt.xlabel("Stát")
plt.show()
```

4. **Počet filmů a seriálů**

Dle grafu došlo k zjištění, že pouze 2 země mají více jak 500

```python
plt.figure(figsize=(8,6))
sns.countplot(x='type', data=df)
plt.title('Distribuce filmů a televizních pořadů')
plt.ylabel("Počet")
plt.xlabel("Typ")
plt.show()
```

5. **Vydání dle roku**

```python
plt.figure(figsize=(10,5))
sns.histplot(df['release_year'], bins=100, kde=False)
plt.title('Distribuce vydání')
plt.ylabel("Počet")
plt.xlabel("Datum vydání")
plt.show()
```
